```python
from sklearn import datasets
import matplotlib.pyplot as plt
from sklearn.cluster import AgglomerativeClustering
```

```python
df = datasets.load_iris()
x = df.data
x[:4] 
```

```python
#Seperar data para entrenamiento y testeo pero no es necesario separarlo por que no hay nuevos datos
#Entonces se separaran los datos
xs = x[:,0]
ys = x[:,1]
plt.scatter(xs,ys)
```

```python
model = AgglomerativeClustering(n_clusters=4) #Por decto ward como criterio de agrupacion
predict = model.fit_predict(x)
plt.scatter(xs,ys, c= model.labels_)
```
